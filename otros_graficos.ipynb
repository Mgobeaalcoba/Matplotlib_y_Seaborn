```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100) # 100 datos aleatorios del 1 al 50
```

```python
plt.hist(data) # Histograma o distribución de frecuancia grafica
plt.show()
```

```python
plt.hist(data, bins=25) # Histograma o distribución de frecuancia grafica
plt.show()
```

```python
plt.hist(data, bins=50, histtype= 'bar') # Histograma o distribución de frecuancia grafica
plt.show()
```

```python
plt.hist(data, bins=50, histtype= 'step') # Histograma o distribución de frecuancia grafica
plt.show()
```

# Otra forma de ver la distribución de nuestros graficos es un box plot o grafico de caja

```python
plt.boxplot(data) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos
plt.show()
```

```python
plt.boxplot(data, vert=False) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos
plt.show()
```

```python
plt.boxplot(data, vert = False, patch_artist = True) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos
plt.show()
```

```python
plt.boxplot(data, vert = False, patch_artist = True, notch = True) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos mas el minimo y el maximo
plt.show()
```

**Un boxplot nos sirve para identificar con facilidad los outliers o datos que desentonan del resto del conjunto.**

```python
data = np.append(data,200)
data
```

```python
plt.boxplot(data, vert = False, patch_artist = True, notch = True) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos
plt.show()
```

```python
# Si tengo outliers y no quiero verlos puedo hacerlo así: 
plt.boxplot(data, vert = False, patch_artist = True, notch = True, showfliers = False) # Grafico de caja que me muesta la mediana y los percentiles 23 y 75 de mis datos
plt.show()
```

### Scatter Plot o Grafico de dispersión

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, s=area) # s sirve para definir el size o tamaño de cada uno de los puntos en función de una tercera variable que en este caso es area
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors) # c sirve para cambiar los colores de los circulos en función de otra varibale que es el color
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='p') # con marker puedo tmb editar el tipo de marca en el scatter plot. En este caso usaré un poligono
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='p', alpha=0.5) # con alpha puedo alterar la transparencia dee los poligonos para hacer que se vean los superpuestos. 
plt.show()
```
