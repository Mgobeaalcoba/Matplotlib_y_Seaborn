**Método Orientado a Objetos**

Pyplot: Rápido, Fácil, Una sola figura.

Object Oriented: Mayor personalización, Más código, Más amigable a multiples diagramas

Con el metodo orientado a objetos de matplotlib puedo trazar en un mismo grafico dos graficas que hagan referencia a distintos ejes verticales. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure() # Creo la figura vacia o lienzo donde dejaré mis graficos
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # Creo mi primer axis y defino la posición inicial del gráfico(dos primeras variables), mas el tamaño en largo y ancho (últimas dos variables)
axes.plot(x,y,'b') # Grafico un plot sobre el axes creado arriba / Atención: En objetos el plot o grafico se hace sobre el axes mientras que en pyplot se hace sobre plt mismo. 
fig.show() # Muestro la figura con show. / Tmb se ejecuta sobre la figura y ya no sobre plt. 
```

```python
# De esta forma puedo incorporar dos graficos de distintos tamaños y formas en un mismo lienzo

fig = plt.figure() #Objeto

axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])    # Axis para el primer objeto
axes2 = fig.add_axes([0.14, 0.65, 0.4, 0.3]) # Axis para el segundo objeto

axes.plot(x,y,'b')   #Plot 1
axes2.plot(y,x,'r')  #Plot 2
axes2.set_facecolor('gray')

fig.show()

# los axis son objetos completamente distintos que reposan sobre un mismo lienzo
```

De esta forma puedo hacer grafico 100% personalizados. 

**Importante: No confundir "Axes" que son los objetos que trabajo dentro de un lienzo con "Axis" que son los ejes de cada uno de los graficos**
