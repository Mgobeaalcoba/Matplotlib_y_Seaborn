```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Defino mi lienzo trabajando con subpolots

fig, axes = plt.subplots()
```

```python
# Hago un grafico simple con mi lienzo definido

fig, axes = plt.subplots()
axes.plot(x,y, 'bo-')
```

```python
# Hago mas de un subplot

fig, axes = plt.subplots(nrows=1, ncols=2) #Defino la cantida de subplots que voy a tener mediante filas y columnas. En este caso 2 subplots
axes[0].plot(x,y, 'bo-') # Defino solo el primer subplot por lo que el segundo queda vacio. 
```

```python
# Hago mas de un subplot

fig, axes = plt.subplots(nrows=1, ncols=2) #Defino la cantida de subplots que voy a tener mediante filas y columnas. En este caso 2 subplots /  Axes en este caso es un array, por lo que debe ser llamado cada axes por su posición.
axes[0].plot(x,y, 'bo-') # Defino el primer subplot
axes[1].plot(y,x, 'rx--') # Defino el segundo subplot
```

```python
# Otra fornma de definir y armar los subplot es mediante una tupla de subplot al momento de definir la variable y luego al personalizarlos. De la siguiente forma: 

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) #Defino la cantida de subplots que voy a tener mediante filas y columnas. En este caso 2 subplots
ax1.plot(x,y, 'bo-') # Defino el primer subplot
ax2.plot(y,x, 'rx--') # Defino el segundo subplot
```

```python
fig, axes = plt.subplots(2, 4) 
axes[0,0].plot(x,y, 'bo-') # Defino el primer subplot / Cuando tenía una sola fila no tenía que especificar fila y columna del subplot. Ahora si debo hacerlo
axes[1,0].plot(y,x, 'rx--') # Defino el segundo subplot
axes[0,1].plot(x,np.cos(x))
axes[0,2].plot(x,np.sin(x))
axes[0,3].plot(x,np.tan(x))
axes[1,1].plot(y,np.cos(y))
axes[1,2].plot(y,np.sin(y))
axes[1,3].plot(y,np.tan(y))

# Complete cada axes de forma individual
```

```python
# Con el segundo metodo que vimos para definir los axes sería así: 

fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4) 
ax1.plot(x,y, 'bo-') # Defino el primer subplot / Cuando tenía una sola fila no tenía que especificar fila y columna del subplot. Ahora si debo hacerlo
ax2.plot(y,x, 'rx--') # Defino el segundo subplot
ax3.plot(x,np.cos(x))
ax4.plot(x,np.sin(x))
ax5.plot(x,np.tan(x))
ax6.plot(y,np.cos(y))
ax7.plot(y,np.sin(y))
ax8.plot(y,np.tan(y))
fig.tight_layout() # Presenta los graficos con espacio suficiente para apreciarlos
fig.show()
```
