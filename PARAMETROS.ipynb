```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn tiene ya pre cargados distintos datasets con los que uno puede "jugar" y practicar graficación.

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips
```

```python
sns.displot(data=tips, x='total_bill') # Grafico de distribución de frecuencias o histograma
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip') 
plt.show()
# En este caso, pese a que le pedí un displot, al reconocer que le pasé mas de una variable para el mismo me va a graficar con un grafico mas adecuado. En este caso un scatter plot
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex') # Con hue le paso un parametro de agrupamiento de los datos. Generalmente se usan variables categoricas 
plt.show()
```

```python
# Tmb funciona con los histogramas el hue
sns.displot(data=tips, x='total_bill', hue='sex') # Con hue le paso un parametro de agrupamiento de los datos. Generalmente se usan variables categoricas 
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind= 'kde') # con kind= puedo seleccionar distintas stilos para el mismo tipo de grafico
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind= 'kde', legend=False) # con legend False apago las etiquetas si es que no quiero mostrarlas
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind= 'kde', legend=False, palette='dark') # con palette puedo cambiar la paleta de colores
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind= 'kde', legend=False, palette='dark', alpha=0.3) # con alpha puedo mostrar una grafico con mayor o menor transparencia
plt.show()
```
