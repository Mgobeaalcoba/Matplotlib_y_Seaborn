```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips.head()
```

```python
# Son dos graficos en uno o tres: Dos dist de frecuencia, una para cada eje mas un scatter plot que muestra la relaci√≥n entre ambas
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# Puedo setear el estilo que quiero ver. Por ejemplo pidiendole por el parametro kind= un histograma
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
# Puedo setear el estilo que quiero ver. Por ejemplo pidiendole por el parametro kind= un kde para los tres
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
# Distintos kinds de jointplot: ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
# Puedo setear el estilo que quiero ver. Por ejemplo pidiendole por el parametro kind= un kde para los tres
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='scatter', marginal_ticks=True) # Le agrega eje de altura a cada grafico marginal
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
              marginal_kws=dict(bins=25, fill=False, multiple='dodge')) # Le agrego parametros adicionales a los graficos marginales
plt.show()
```

```python
# El pairplot me muestra la distribuci√≥n y relacion de todas las variables numericas que yo tenga en mi dataset
sns.pairplot(data=tips)
plt.show()
```

```python
# El pairplot me muestra la distribuci√≥n y relacion de todas las variables numericas que yo tenga en mi dataset
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True) # Con corner=True apago directamente los graficos repetidos
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='scatter') # puedo cambiar tmb los tipos de graficos con kind=
plt.show()
```
