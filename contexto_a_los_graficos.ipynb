```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows= 2, ncols= 2)
```

```python
fig, axes = plt.subplots(nrows= 1, ncols= 2, )
axes[0].plot(x,y)
axes[1].plot(y,x)
fig.show()
fig.tight_layout() # Si uso este comando no puedo usar el figsize=(7,5) dentro de la definición de la figura
# Solo con esto las graficas son escuetas. No tienen referencias ni contexto. 
axes[0].set_title('Relación X - Y') # Pongo titulo al axes 0
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X  ')
```

```python
fig, axes = plt.subplots(nrows= 1, ncols= 2, figsize=(7,5) )
axes[0].plot(x,y)
axes[1].plot(y,x)
fig.show()
#fig.tight_layout() # Si uso este comando no puedo usar el figsize=(7,5) dentro de la definición de la figura
# Solo con esto las graficas son escuetas. No tienen referencias ni contexto. 
axes[0].set_title('Relación X - Y') # Pongo titulo al axes 0
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X  ')
```

```python
# Manejo de labeels en los graficos: los defino en el plot pero para que se muestren debo luego sumar al axes en cuestión la función legend()
fig, axes = plt.subplots(nrows= 1, ncols= 2, figsize=(7,5) )
axes[0].plot(x,y, 'ro-', label= '$Sin(x)$') # Rodeando de signos pesos un nombre matematico pasa a notación matematica
axes[1].plot(y,x, label= '$Cos(y)$')
axes[0].legend()
axes[1].legend()
fig.show()
axes[0].set_title('Relación X - Y') # Pongo titulo al axes 0
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
# Las labels tmb se pueden usar en pyplot por fuera del metodo orientado a objetos...
plt.plot(x,y, 'ro-', label='$sin(x)$', )
plt.title('Relación X-Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.8,1)) # Puedo seleccionar la ubicación de mi label con loc / con bbox_to_anchor=(0.5,0.1) puedo sacar la label del plot / Por default se acomoda de la mejor manera
plt.show()
```

# Matlab : Laboratorio de estilos de matplotlib

Se usa como tercer parametro en los plot de matplotlib orientado a objetos. 

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x+1, 'r--') # Puedo poner en mi grafico todas las lineas que desee separandolas por color. 
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
fig.show()
```

# Tmb se puede hacer con pyplot de la siguiente forma: 

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x+1, color = 'green', alpha = 0.8, linewidth = 8, linestyle = '-', marker = 'o', markersize= 20, markerfacecolor = '#FFCCFF') # Puedo poner en mi grafico todas las lineas que desee separandolas por color. 
ax.plot(x, x+2, color = 'blue', linewidth = 5, linestyle = '--', marker = 'x')
ax.plot(x, x+3, color = '#66FF66', linewidth = 2, linestyle = 'dashed', marker = 'v') # Al usar pyplot para definir colores puedo usar cualquier variante de la paleta RPG
ax.plot(x, x+4, color = '#CCFFCC', linewidth = 10, linestyle = ':', marker = 'P')
fig.show()
# Puedo definir tmb la transparencia de la linea con 'alpha', el grosor de la linea con 'linewidth' y el estilo de la linea con 'linestyle', tmb puedo definir los marcadores con 'marker', 
# el tamaño de los markers con markersize y el color de los marker, si deseo separarlo del color de la linea con markerfacecolor
```
