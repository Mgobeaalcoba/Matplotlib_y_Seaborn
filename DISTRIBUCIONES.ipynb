```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips.head()
```

```python
sns.histplot(data=tips, x='tip') # Histograma de Seaborn
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=50) # Puedo cambiar los bins como matplotlib
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=50, cumulative=True) # con cumulative puedo ir viendo el acumulado de la distribución
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=50, cumulative=False, hue='sex') # con hue abro los datos por una variable generalmente categorica.
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=50, cumulative=False, hue='sex', stat='frequency') # con stat puedo mostrar el histograma en frecuencia, probabilidad, etc
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=50, cumulative=False, hue='sex', stat='probability') # con stat puedo mostrar el histograma en frecuencia, probabilidad, etc
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='percent') # con stat puedo mostrar el histograma en frecuencia, probabilidad, etc
plt.show()
```

```python
# ['layer', 'stack', 'fill', 'dodge']
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='layer') # Con multiple puedo seleccionar la forma de apilamiento o presentación del aperturado del hue
plt.show()
```

```python
# ['layer', 'stack', 'fill', 'dodge']
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='stack') # Con multiple puedo seleccionar la forma de apilamiento o presentación del aperturado del hue
plt.show()
```

```python
# ['layer', 'stack', 'fill', 'dodge']
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge') # Con multiple puedo seleccionar la forma de apilamiento o presentación del aperturado del hue
plt.show()
# Acá no serían apiladas las barras sino una al lado de la otra. 
```

```python
# ['layer', 'stack', 'fill', 'dodge']
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='fill') # Con multiple puedo seleccionar la forma de apilamiento o presentación del aperturado del hue
plt.show()
# Barras apiladas completas
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex') # histograma con lineas
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True) # histograma con lineas
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True) # Pinta las areas bajo las lineas para facilitar su lectura
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=2) # Para suavizar las lineas podemos usar bw_adjust
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex') # Es una grafico escalonado y acumulativo
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count') # Es una grafico escalonado y acumulativo
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex',kind='hist', multiple='stack')
plt.show()
```
